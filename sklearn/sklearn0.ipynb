```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Create a pipeline object
pipe = make_pipeline(
    StandardScaler(),
    LogisticRegression()
)

# Split the data into train and test
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Fit the whole pipeline
pipe.fit(X_train, y_train)

# Assess performance
accuracy_score(pipe.predict(X_test), y_test)
```

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate

# Simulate data
X, y = make_regression(n_samples = 1000, random_state=42)

# Set up linear model object
lr = LinearRegression()

# Conduct cross validation, which includes model fitting
result = cross_validate(lr, X, y)

# Examine test_score
result['test_score']
```

```python
from sklearn.datasets import make_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from scipy.stats import randint

# Create a synthetic dataset
X, y = make_regression(n_samples=20640,
                       n_features=8,
                       noise=0.1,
                       random_state=42)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Define the hyperparameter space
param_distributions = {'n_estimators': randint(1, 5),
                       'max_depth': randint(5, 10)}

# Create a RandomSearchCV object and fit with data
search = RandomizedSearchCV(estimator=RandomForestRegressor(random_state=42),
                        n_iter=5,
                        param_distributions=param_distributions,
                        random_state=42)
search.fit(X_train, y_train)

# Get best hyperparameters
search.best_params_
# {'max_depth': 7, 'n_estimators': 4}

# search object can then be used in making predictions
search.score(X_test, y_test)
# 0.7925192997818057
```
