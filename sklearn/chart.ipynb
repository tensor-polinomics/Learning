```python
def my_plotter(ax, x, y, param_dict):
    out = ax.plot(x, y, **param_dict)
    return out
```

```python
x = np.linspace(0, 10, 100)
y1, y2, y3 = np.random.randn(3, 100)

fig, ax = plt.subplots(1, 3, figsize=(11,3))
params = dict(color='tab:blue', linestyle='--', marker='o')
for a, y in zip(ax, [y1, y2, y3]):
    my_plotter(a, x, y, params)
    a.set_ylabel('y')
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(5,3), facecolor='lightblue')
x = np.arange(len(y1))
ax.plot(x, y1.cumsum(), linewidth=2)
l, = ax.plot(x, y2.cumsum(), color='orange')
l.set_linestyle(':')
ax.set_facecolor('lightblue')
ax.set_yscale('log')
```

```python
y4 = y3 + 100
fig, ax1 = plt.subplots(figsize=(5,3))
ax2 = ax1.twinx()
ax1.plot(x, y1)
ax2.plot(x, y4, color='orange')
ax2.set_ylabel('y4')
```

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(1000)
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(x, 50, density=True, facecolor='lightgray');
ax.text(75, 0.025, r'$\mu=115, \ \sigma=15$')
ax.set_xlabel("x-Axis", fontsize=10, color='red')
ax.grid(True)
ax.annotate('Peak', xy=(115, 0.03), xytext=(130, 0.025),
            arrowprops=dict(facecolor='blue', shrink=0.01))
```
