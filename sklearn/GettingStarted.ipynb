```python
air_quality_no2 = pd.read_csv("data/air_quality_no2_long.csv")
air_quality_no2.head()
```

```python
# rename date.utc to datetime
# as of now "datetime" is "object", not a datetime object
air_quality_no2 = air_quality_no2.rename(columns={"date.utc": "datetime"})
air_quality_no2.info()
```

```python
air_quality_no2["datetime"] = pd.to_datetime(air_quality_no2["datetime"])
air_quality_no2.info()
# Now, datetime is a datetime object called pd.Timestamp
```

```python
temp = pd.read_csv("data/air_quality_no2_long.csv", parse_dates=["date.utc"])
temp.info()
```

```python
air_quality_no2["datetime"].min(), air_quality_no2["datetime"].max()
```

```python
# Subtracting two Timestamp objects returns a timedelta object
air_quality_no2["datetime"].max() - air_quality_no2["datetime"].min()
```

```python
air_quality_no2['month'] = air_quality_no2["datetime"].dt.month
air_quality_no2[air_quality_no2['month'] == 6]
```

```python
# Compute the average daily value, for each location
# note: monday=0
air_quality_no2.groupby([air_quality_no2["datetime"].dt.weekday, "location"])["value"].mean()
```

```python
# Compute the average value by hour of the day
fig, axs = plt.subplots(figsize=(8,4))
air_quality_no2.groupby(air_quality_no2["datetime"].dt.hour)["value"].mean().plot(
    kind="bar", rot=0, ax=axs)
```
