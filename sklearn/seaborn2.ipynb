```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(fmri, kind="line", x="timepoint", y="signal",
            row="event", col="region", height=3)
plt.savefig("data/fmri_grid.png")
```

```python
sns.relplot(data=fmri.query("region=='frontal'"), kind="line",
            x="timepoint", y="signal", hue="event",
            col="subject", col_wrap=4, height=3, aspect=.75)
plt.savefig("data/fmri_many.png")
```

```python
sns.lmplot(tips, x="total_bill", y="tip",
           lowess=True, height=3, aspect=1.5,
           line_kws={"color": "red"})
plt.savefig("data/tips_lowess2.png")
```

```python
sns.lmplot(tips, x="total_bill", y="tip", hue="smoker",
           lowess=True, height=3, aspect=1.5,
           scatter_kws={"alpha": 0.5, "s": 10})
plt.savefig("data/tips_by_smoker.png")
```

```python
sns.lmplot(tips, x="total_bill", y="tip", col="time", hue="smoker",
           lowess=True, height=3, aspect=1,
           scatter_kws={"alpha": 0.5, "s": 10})
plt.savefig("data/tips_by_smoker_time.png")
```

```python
sns.jointplot(tips, x="total_bill", y="tip", kind="reg")
plt.savefig("data/tips_jointplot_reg.png")
```

```python
sns.pairplot(tips, x_vars=["total_bill", "size"], y_vars=["tip"],
             height=4, aspect=0.8, kind="reg")
plt.savefig("data/tips_pairs_reg.png")
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(sns.histplot, "tip")
plt.savefig("data/tips_facet2.png")
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(sns.scatterplot, "total_bill", "tip", alpha=.7)
g.add_legend()
plt.savefig("data/tips_facet3.png")
```

```python
g = sns.FacetGrid(tips, row="smoker", col="time", margin_titles=True)
g.map(sns.regplot, "size", "total_bill", ci=False)
plt.savefig("data/tips_facet4.png")
```

```python
tips.day.value_counts()
```

```python
g = sns.FacetGrid(tips, col="day", height=4, aspect=.5)
g.map(sns.barplot, "sex", "total_bill", order=["Male", "Female"])
plt.savefig("data/tips_facet5.png")
```

```python
g = sns.FacetGrid(tips, row="sex", col="smoker", margin_titles=True, height=2.5)
g.map(sns.scatterplot, "total_bill", "tip")
g.set_axis_labels("Total Bill in USD", "Tip")
g.set(xticks=[10, 20, 30, 40, 50])
g.figure.subplots_adjust(wspace=0.02, hspace=0.02)
plt.savefig("data/tips_facet6.png")
```

```python
iris = sns.load_dataset("iris")
```

```python
sns.pairplot(iris)
plt.savefig("data/iris_pair_default.png")
```

```python
g = sns.PairGrid(iris)
g.map(sns.scatterplot)
plt.savefig("data/iris_pair_scatter.png")
```

```python
g = sns.PairGrid(iris)
g.map_offdiag(sns.scatterplot)
g.map_diag(sns.histplot)
plt.savefig("data/iris_pair_scahist.png")
```

```python
g = sns.PairGrid(iris, hue="species")
g.map_offdiag(sns.scatterplot)
g.map_diag(sns.kdeplot)
g.add_legend()
plt.savefig("data/iris_pair_scakde.png")
```

```python
g = sns.PairGrid(iris)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot)
plt.savefig("data/iris_pair_uplo.png")
```

```python
g = sns.PairGrid(tips, y_vars=["tip"], x_vars=["total_bill", "size"])
g.map(sns.regplot, scatter_kws={"alpha": 0.5, "s": 10}, 
      line_kws={"color": "red"})
plt.savefig("data/iris_pair_reg.png")
```
